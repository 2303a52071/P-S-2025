```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df= pd.read_csv('/content/student_data_250.csv')
df.head()
```

```python
#  Histogram Analysis:
# to Plot a histogram to visualize the distribution of GPA values in the dataset.
#  Use appropriate bin sizes and axis labels.
# oAnalyze the shape of the distribution (e.g., normal, skewed)

# Assuming 'GPA' is the column name containing GPA values in your DataFrame
plt.figure(figsize=(8, 6))
sns.histplot(df['GPA'], bins=10, kde=True)  # Adjust the number of bins as needed
plt.xlabel('GPA')
plt.ylabel('Frequency')
plt.title('Distribution of GPA Values')
plt.show()

# Analyze the shape of the distribution
# You can visually inspect the histogram to determine the shape.
# If it's roughly symmetrical and bell-shaped, it's approximately normal.
# If it's skewed to the right (tail on the right), it's positively skewed.
# If it's skewed to the left (tail on the left), it's negatively skewed.

#Kernel Density Estimation (KDE)

#kde=True enables the plotting of a
#Kernel Density Estimate (KDE) along with the histogram.
#What is KDE?: KDE is a way to estimate the probability density function of a continuous
#random variable. In simpler terms, it creates a smooth curve that represents
#the distribution of the data.
```

```python
# prompt: Scatterplot Exploration:
# Create a scatterplot of GPA vs. Attendance Rate (%).
# Interpret whether there is a correlation between attendance and academic
# performance.
# Add axis labels and a trendline to the plot.

# Scatterplot Exploration:
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Attendance Rate (%)', y='GPA', data=df)
plt.xlabel('Attendance Rate (%)')
plt.ylabel('GPA')
plt.title('GPA vs. Attendance Rate')

# Add a trendline
sns.regplot(x='Attendance Rate (%)', y='GPA', data=df, scatter=False, color='red')

plt.show()

# Interpret the correlation
# A positive correlation suggests that as attendance rate increases, GPA tends to increase.
# A negative correlation suggests that as attendance rate increases, GPA tends to decrease.
# A weak correlation suggests little to no relationship between attendance and GPA.
# Visually inspect the scatterplot and trendline to determine the correlation.
```

```python
# prompt: Bar Chart Visualization:
# to Generate a bar chart showing the number of students per Major.
# to Analyze which majors have the highest student enrollment.
# to Ensure the chart includes appropriate titles and axis labels.


# Bar Chart Visualization
plt.figure(figsize=(12, 6))
sns.countplot(x='Major', data=df)
plt.xlabel('Major')
plt.ylabel('Number of Students')
plt.title('Number of Students per Major')
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for readability
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()

# Analyze majors with the highest enrollment
major_counts = df['Major'].value_counts()
print("\nMajors with the highest student enrollment:")
print(major_counts.head(5)) # Print top 5 majors
```

```python
 # prompt: Grouped Bar Chart:
# to Create a grouped bar chart comparing the number of students with and without
# extracurricular activities across different Graduation Years.
# to Discuss if there is any visible trend in extracurricular involvement over the years.


# Assuming your dataframe is named 'df' and it has columns 'Grad Year', 'Extracurricular Activities'
plt.figure(figsize=(10, 6))
# Removed the trailing whitespace in the 'hue' parameter
sns.countplot(x='Graduation Year', hue='Extracurricular Activities', data=df)
plt.xlabel('Graduation Year')
plt.ylabel('Number of Students')
plt.title('Extracurricular Activities by Graduation Year')
plt.legend(title='Extracurricular Activities')
plt.show()

# Trend analysis
# Observe the grouped bars for each graduation year.
# Look for patterns in the heights of the bars representing students with and without extracurricular activities.
# If the bars for "Yes" consistently increase or decrease over the years, there's a trend.
# If there's no clear pattern, there may be no discernible trend.
# You can quantify the trend by calculating the percentage of students with extracurricular activities each year.
```
