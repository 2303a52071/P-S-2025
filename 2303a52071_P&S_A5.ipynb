```python
# prompt: Compute and visualize the PMF and CDF to analyze the probability of receiving a
# specific number of calls on a given day.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def generate_call_data(N):
    """
    Generates a dataset representing the number of calls received per day for N days.

    Args:
        N: The number of days for which to generate data.

    Returns:
        A pandas DataFrame with a 'calls' column representing the number of calls each day.
    """
    calls = np.random.randint(0, 8, size=N) # Generate random integers between 0 and 7 (inclusive)
    df = pd.DataFrame({'calls': calls})
    return df

# Example usage:
N = 100 # For 100 days
call_data = generate_call_data(N)

# Calculate PMF
call_counts = call_data['calls'].value_counts().sort_index()
pmf = call_counts / len(call_data)

# Calculate CDF
cdf = pmf.cumsum()

# Plotting PMF
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.bar(pmf.index, pmf.values)
plt.xlabel('Number of Calls')
plt.ylabel('Probability')
plt.title('Probability Mass Function (PMF)')
plt.xticks(range(0,8))
plt.grid(True)


# Plotting CDF
plt.subplot(1, 2, 2)
plt.plot(cdf.index, cdf.values, marker='o', linestyle='-')
plt.xlabel('Number of Calls')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Distribution Function (CDF)')
plt.xticks(range(0,8))
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# prompt: A hospital receives emergency cases daily, and there is a 25% probability that a randomly
# arriving paƟent requires immediate surgery. Suppose N paƟents visit the emergency
# department in a day (N ≥ 100):
# i) What is the probability that exactly 20 paƟents need surgery
# with graph i want, i want bar graph

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import binom

def calculate_binomial_probability(n, k, p):
    """
    Calculates the probability of exactly k successes in n trials using the binomial distribution.

    Args:
        n: The number of trials (patients).
        k: The number of successes (patients needing surgery).
        p: The probability of success (probability of a patient needing surgery).

    Returns:
        The probability of exactly k successes in n trials.
    """
    return binom.pmf(k, n, p)

# Given parameters
N = 100  # Number of patients
k = 20   # Number of patients needing surgery
p = 0.25 # Probability of a patient needing surgery

# Calculate the probability
probability = calculate_binomial_probability(N, k, p)
print(f"The probability of exactly {k} patients needing surgery out of {N} is: {probability:.6f}")

# Generate a range of k values for plotting
k_values = np.arange(0, N+1)
probabilities = [calculate_binomial_probability(N, i, p) for i in k_values]

# Plotting the PMF
plt.figure(figsize=(10, 6))
plt.bar(k_values, probabilities, color='skyblue')
plt.xlabel('Number of Patients Needing Surgery (k)')
plt.ylabel('Probability')
plt.title('Probability Mass Function (PMF) of Patients Needing Surgery')
plt.grid(axis='y', alpha=0.75)
plt.xlim(0, N)
plt.show()
```
